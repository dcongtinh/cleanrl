```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated data
x = np.linspace(0, 1, 100)
y1 = 300 * (1 - np.exp(-3 * x)) + np.random.normal(0, 10, 100)
y2 = 180 * (1 - np.exp(-2.5 * x)) + np.random.normal(0, 8, 100)

# Simulated std for shading
std1 = 40 + 5 * np.random.rand(100)
std2 = 30 + 4 * np.random.rand(100)

# Plot setup
plt.figure(figsize=(5, 3.5))

# First plot
plt.plot(x, y1, color='firebrick', linewidth=2)
plt.fill_between(x, y1 - std1, y1 + std1, color='firebrick', alpha=0.2)

# Second plot
plt.plot(x, y2, color='cornflowerblue', linewidth=2)
plt.fill_between(x, y2 - std2, y2 + std2, color='cornflowerblue', alpha=0.2)

# Title and labels
plt.title("acrobot-swingup", fontsize=12, color='gray')
plt.xlabel("")
plt.ylabel("Average Episode Score", fontsize=10, color='gray')

# Axis style
plt.tick_params(axis='x', colors='gray')
plt.tick_params(axis='y', colors='gray')
plt.grid(True, alpha=0.3)

# Remove top/right borders
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Tight layout and show
plt.tight_layout()
plt.show()
```
